K mean Clustering

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/clustering_dataset.csv')
df.head(6)
```

```python
# Features
X = df[['Feature1', 'Feature2']]

# Initialize KMeans with k=3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model
kmeans.fit(X)

# Assign clusters
df['Cluster'] = kmeans.labels_
```

```python
# Plotting clusters
plt.figure(figsize=(6,4))
plt.scatter(df['Feature1'], df['Feature2'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.title('K-Means Clustering')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(X, kmeans.labels_)
print("Silhouette Score:", score)
```

```python
new_points = [[3, 4], [10, 10], [1, 0]]
predicted_clusters = kmeans.predict(new_points)
print(predicted_clusters)
```
