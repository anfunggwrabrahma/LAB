```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/linear_dataset.csv')
df
```

```python
# Display basic statistics
print("Dataset statistics:")
print(df.describe())

# Plot data distribution
plt.figure(figsize=(6,4))
plt.scatter(df['Hours_Studied'], df['Test_Score'], color='green')
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Scatter Plot: Hours Studied vs Test Score')
plt.grid(True)
plt.show()
```

```python
# Features and target
X = df[['Hours_Studied']]
y = df['Test_Score']

# Train linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Regression coefficients
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
```

```python
# Plotting
plt.scatter(X, y, color='blue', label='Actual Scores')
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Linear Regression: Hours Studied vs Test Score')
plt.legend()
plt.grid()
plt.show()
```

```python
hrs_studied = 2
pred_score = model.predict([[hrs_studied]])
print(f"Predicted test score for {hrs_studied} hours studied: {pred_score[0]}")
```
