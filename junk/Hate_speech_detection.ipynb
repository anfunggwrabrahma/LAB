```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('indian')  # Download the Indian language corpus (includes Bodo stopwords)

def text_cleaning(text):
    # Remove special characters and digits
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    cleaned_text = re.sub(r'\d+', '', cleaned_text)

    # Convert to lowercase
    cleaned_text = cleaned_text.lower()

    return cleaned_text

def tokenization(text):
    # Tokenize the text into individual words
    tokens = word_tokenize(text)

    return tokens

def remove_stopwords(tokens):
    # Get the set of stopwords for the Bodo language
    stop_words = set(stopwords.words('indian') + ['थों','मोन'
'हागौ', 'था', 'जा','नाय'])

    # Remove stopwords from the tokens
    filtered_tokens = [token for token in tokens if token not in stop_words]

    return filtered_tokens

# Load the CSV file into a Pandas DataFrame
df = pd.read_csv('/content/drive/MyDrive/side_project/training.csv')

# Assuming 'text_column' is the name of the column that contains the text data
df['cleaned_text'] = df['text'].apply(text_cleaning)
df['tokens'] = df['cleaned_text'].apply(tokenization)
df['filtered_tokens'] = df['tokens'].apply(remove_stopwords)

# Print the preprocessed data
print(df[['text', 'cleaned_text', 'filtered_tokens']])
```
