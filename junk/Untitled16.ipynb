```python
from sacremoses import MosesTokenizer

# Initialize the Moses tokenizer
tokenizer = MosesTokenizer()

# Input English text
english_text = "This is a sample sentence for tokenization."

# Tokenize the English text using Moses tokenizer
tokens = tokenizer.tokenize(english_text)

# Print the tokens
print(tokens)
```
