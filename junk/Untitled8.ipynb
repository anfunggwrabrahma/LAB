```python
import csv
csv.field_size_limit(100000000)  # set the maximum field size to 100 MB
with open('/content/drive/MyDrive/merged_file.csv') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    data = list(reader)
```

```python
with open('/content/drive/MyDrive/merged_file.txt', 'w') as outfile:
    for row in data:
        if len(row) > 1:
            source = row[0]
            target = row[1]
            outfile.write(source + '\t' + target + '\n')
```

```python
!onmt_preprocess -train_src /content/drive/MyDrive/merged_file.txt \
                 -train_tgt /content/drive/MyDrive/merged_file.txt \
                 -save_data /content/drive/MyDrive/preprocessed_data \
                 -src_vocab_size 50000 \
                 -tgt_vocab_size 50000
```

```python
!onmt_train -data /content/drive/MyDrive/preprocessed_data \
            -save_model /content/drive/MyDrive/model \
            -world_size 1 \
            -gpu_ranks 0 \
            -backend nccl \
            -train_steps 10000 \
            -valid_steps 500 \
            -save_checkpoint_steps 1000 \
            -batch_size 64 \
            -encoder_type brnn \
            -enc_layers 2 \
            -dec_layers 2 \
            -rnn_size 512 \
            -dropout 0.1 \
            -global_attention general \
            -optim adam \
            -learning_rate 0.001 \
            -learning_rate_decay 0.5 \
            -start_decay_steps 10000 \
            -decay_steps 5000 \
            -log_file /content/drive/MyDrive/train.log
```
