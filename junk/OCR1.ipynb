```python
import os
# Paths to input and output directories where the dataset
and output files will be saved in png and txt

input_dir = "/content/TestOCR"
output_dir = "/content/OutputOCR"
os.makedirs(output_dir, exist_ok=True)

for file_name in os.listdir(input_dir):
    if file_name.lower().endswith(".png"):
        input_path = os.path.join(input_dir, file_name)
        output_base = "output_" + os.path.splitext(file_name)[0]
        output_path = os.path.join(output_dir, output_base)

        # Run Tesseract OCR with Hindi + English
        !tesseract "{input_path}" "{output_path}" -l eng+hin --oem 1 --psm 3

        print(f"OCR saved as: {output_base}.txt")
```
